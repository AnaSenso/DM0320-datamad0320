```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.datasets import make_blobs
from yellowbrick.cluster import KElbowVisualizer
from sklearn.cluster import AgglomerativeClustering
```

```python
from sklearn import datasets

data = datasets.load_wine()

X = pd.DataFrame(data["data"], columns=data["feature_names"])
y = pd.Series(data["target"])
```

```python
X.head()
```

## 1. Train a KMeans clustering model on the data set using 8 clusters and compute the silhouette score for the model.

```python
model = KMeans(n_clusters=8)
labels = model.fit_predict(X)
```

```python
silhouette_8 = silhouette_score(X,labels)
silhouette_8
```

## 2. Train a KMeans clustering model on the data set using 5 clusters and compute the silhouette score for the model.

```python
model = KMeans(n_clusters=5)
labels = model.fit_predict(X)
```

```python
silhouette_5 = silhouette_score(X,labels)
silhouette_5
```

## 3. Train a KMeans clustering model on the data set using 3 clusters and compute the silhouette score for the model.

```python
model = KMeans(n_clusters=3)
labels = model.fit_predict(X)
```

```python
silhouette_3 = silhouette_score(X,labels)
silhouette_3
```

## 4. Use elbow curve visualizations to see if you can determine the best number of clusters to use.

The Yellowbrick library has 3 metrics that you can plot using the `metric` parameter:

- **distortion**: mean sum of squared distances to centers
- **silhouette**: mean ratio of intra-cluster and nearest-cluster distance
- **calinski_harabaz**: ratio of within to between cluster dispersion

```python
visualizer = KElbowVisualizer(model, k=(4,12), metric="silhouette", timings=False)
visualizer.fit(X)   
```

## 5. Try performing the same elbow tests with an AgglomerativeClustering model and compare the results you get to the KMeans results.

```python
aggmodel = AgglomerativeClustering(n_clusters=8)
agglabels = aggmodel.fit_predict(X)
```

```python
visualizer = KElbowVisualizer(aggmodel, k=(4,12), metric="silhouette", timings=False)
visualizer.fit(X) 
```

## 6. Create and plot a scatter matrix showing how the clusters are grouped across all the different combinations of variables in the data.

Use the model and number of clusters that returned the best result above.

```python
model_4 = KMeans(n_clusters = 4).fit(X)
labels = model_4.fit_predict(X)
X['label'] = model_4.labels_

plt.subplot(2,2,1)
plt.scatter(X.malic_acid,X.alcohol,c=X.label)
plt.xlabel('malic_acid')
plt.ylabel('alcohol')

plt.subplot(2,2,2)
plt.scatter(X.ash,X.alcohol,c=X.label)
plt.xlabel('ash')
plt.ylabel('alcohol')

plt.subplot(2,2,3)
plt.scatter(X.hue,X.color_intensity,c=X.label)
plt.xlabel('hue')
plt.ylabel('color_intensity')

plt.subplot(2,2,4)
plt.scatter(X.magnesium,X.total_phenols,c=X.label)
plt.xlabel('magnesium')
plt.ylabel('total_phenols')
plt.tight_layout()
```
